Pregunta 1: ¿Cuál es la relación entre la edad de un jugador y su valor de mercado? Presentar un análisis detallado de la relación entre la edad y el valor de mercado, incluyendo visualizaciones y resultados de la regresión lineal.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos_limpios.csv')

# Crear los arrays de variables predictoras y objetivo
X = data['Age'].values.reshape(-1, 1)
y = data['Value'].values

# Ajustar el modelo de regresión lineal
regresion = LinearRegression()
regresion.fit(X, y)

# Obtener los coeficientes de intercepción, pendiente y determinación (R^2)
intercepcion = regresion.intercept_
pendiente = regresion.coef_[0]
r2 = regresion.score(X, y)

# Imprimir los resultados
print("Coeficiente de intercepción:", intercepcion)
print("Coeficiente de pendiente:", pendiente)
print("Coeficiente de determinación (R^2):", r2)

# Graficar los resultados
plt.scatter(X, y)
plt.plot(X, regresion.predict(X), color='red')
plt.title('Relación entre la edad y el valor de mercado')
plt.xlabel('Edad')
plt.ylabel('Valor de mercado')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos_limpios.csv')

# Seleccionar las columnas de habilidades técnicas y valor de mercado
columnas_habilidades = ['Crossing', 'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', 'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control']
columnas_valor_mercado = ['Value']

# Calcular el número total de gráficas y ajustar el número de filas y columnas
num_graficas = len(columnas_habilidades)
num_filas = 3
num_columnas = 4

# Crear la figura y los subplots
fig, axs = plt.subplots(num_filas, num_columnas, figsize=(12, 12))

# Inicializar el contador de gráficas
contador_graficas = 0

# Generar gráficas de dispersión para comparar el valor de mercado con cada habilidad técnica
for fila in range(num_filas):
    for columna in range(num_columnas):
        if contador_graficas < num_graficas:
            habilidad = columnas_habilidades[contador_graficas]
            axs[fila, columna].scatter(data[habilidad], data['Value'])
            axs[fila, columna].set_title(f'Valor de mercado vs {habilidad}')
            axs[fila, columna].set_xlabel(habilidad)
            axs[fila, columna].set_ylabel('Valor de mercado')
            contador_graficas += 1
        else:
            # Si se supera el número total de gráficas, ocultar el subplot vacío
            axs[fila, columna].axis('off')

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

Pregunta 9: ¿Hay una relación entre la altura y el peso de los jugadores y su valor de mercado? Examinar si las características físicas influyen en el valor de mercado.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos_limpios.csv')

# Seleccionar las columnas de altura, peso y valor de mercado
columnas_fisicas = ['Height', 'Weight']
columna_valor_mercado = 'Value'

# Crear un nuevo dataframe con las columnas seleccionadas
data_fisicas = data[columnas_fisicas + [columna_valor_mercado]]

# Eliminar el carácter adicional al final del valor de altura
data_fisicas['Height'] = data_fisicas['Height'].apply(lambda x: int(x.split("'")[0]) * 30.48 + int(x.split("'")[1].split('"')[0]) * 2.54)

# Graficar la relación entre la altura, peso y valor de mercado
sns.pairplot(data_fisicas)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos_limpios.csv')

# Convertir la altura de pies y pulgadas a centímetros
def convertir_altura(altura):
    partes = altura.split("'")
    pies = int(partes[0])
    pulgadas = 0
    if len(partes) > 1:
        pulgadas = int(partes[1].replace('"', ''))
    return pies * 30.48 + pulgadas * 2.54

data['Height'] = data['Height'].apply(convertir_altura)

# Convertir el peso de libras a kilogramos
data['Weight'] = data['Weight'].str.replace(' lbs', '')
data['Weight'] = data['Weight'].str.extract('(\d+.\d+|\d+)').astype(float) * 0.453592

# Definir los rangos de peso y estatura
peso_ranges = [(40, 60), (60, 80), (80, 100)]
altura_ranges = [(160, 170), (170, 180), (180, 190)]

# Calcular el valor promedio por grupo
data['Peso Category'] = pd.cut(data['Weight'], bins=[range[0] for range in peso_ranges] + [peso_ranges[-1][1]])
data['Altura Category'] = pd.cut(data['Height'], bins=[range[0] for range in altura_ranges] + [altura_ranges[-1][1]])
valor_promedio_por_grupo = data.groupby(['Peso Category', 'Altura Category'])['Value'].mean().reset_index()

# Graficar los valores promedio por grupo
plt.figure(figsize=(10, 6))
sns.heatmap(valor_promedio_por_grupo.pivot('Peso Category', 'Altura Category', 'Value'), annot=True, cmap='coolwarm')
plt.title('Valor promedio de mercado por rango de peso y estatura')
plt.xlabel('Altura Category')
plt.ylabel('Peso Category')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos_limpios.csv')

# Seleccionar las columnas de altura, peso y valor de mercado
columnas_fisicas = ['Height', 'Weight']
columna_valor_mercado = 'Value'

# Convertir la altura de pies y pulgadas a centímetros
data['Height'] = data['Height'].apply(lambda x: int(x.split("'")[0]) * 30.48 + int(x.split("'")[1].replace("\"", "")) * 2.54)

# Convertir el peso de libras a kilogramos
data['Weight'] = data['Weight'].apply(lambda x: int(x.split("lbs")[0]) * 0.453592)

# Crear un nuevo dataframe con las columnas seleccionadas
data_fisicas = data[columnas_fisicas + [columna_valor_mercado]]

# Graficar la relación entre la altura, peso y valor de mercado
sns.pairplot(data_fisicas)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos_limpios.csv')

# Seleccionar las columnas de edad, potencial y valor de mercado
columnas_edad_potencial = ['Age', 'POT']
columna_valor_mercado = 'Value'

# Crear un nuevo dataframe con las columnas seleccionadas
data_edad_potencial = data[columnas_edad_potencial + [columna_valor_mercado]]

# Crear una columna para categorías de edad
data_edad_potencial['Edad Category'] = pd.cut(data_edad_potencial['Age'], bins=[0, 20, 25, 30, 35, 40], labels=['<20', '20-25', '26-30', '31-35', '36+'])

# Calcular el promedio del potencial y valor de mercado por categoría de edad
promedio_potencial = data_edad_potencial.groupby('Edad Category')['POT'].mean()
promedio_valor_mercado = data_edad_potencial.groupby('Edad Category')['Value'].mean()

# Graficar el promedio del potencial y valor de mercado por categoría de edad en un gráfico de barras apiladas
plt.figure(figsize=(10, 6))
plt.bar(promedio_potencial.index, promedio_potencial, label='Potencial', color='blue')
plt.bar(promedio_valor_mercado.index, promedio_valor_mercado, label='Valor de mercado', color='orange')
plt.xlabel('Categoría de Edad')
plt.ylabel('Promedio')
plt.title('Promedio de potencial y valor de mercado por categoría de edad')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Cargar los datos desde el archivo CSV
data = pd.read_csv('datos_limpios.csv')

# Preprocesar la columna 'Height_cm'
data['Height_cm'] = data['Height'].apply(lambda x: int(x.split("'")[0]) * 30.48 + int(x.split("'")[1].replace('"', '')) * 2.54)

# Preprocesar la columna 'Weight_kg'
data['Weight_kg'] = data['Weight'].apply(lambda x: int(x.replace('lbs', '')) * 0.453592)

# Definir las características y el valor objetivo
columnas_estadisticas = ['Crossing', 'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys', 'Dribbling', 'Curve',
                         'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility',
                         'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots',
                         'Height_cm', 'Weight_kg']
columna_valor = 'Value'

# Crear una lista de edades para la predicción
edades_prediccion = range(20, 36)  # Edades hasta los 35 años

# Crear un DataFrame para almacenar los resultados de la predicción
resultados_prediccion = pd.DataFrame({'Edad': edades_prediccion, 'Valor': np.nan})

# Obtener las estadísticas promedio por edad
estadisticas_promedio = data.groupby('Age')[columnas_estadisticas].mean()

# Definir las características iniciales
estadisticas_actuales = np.array([60, 65, 65, 60, 60, 60, 60, 65, 60, 70, 60, 65, 70, 60, 70, 65, 75, 65, 70, 60, 170, 75], dtype=np.float64)

# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Iterar por cada edad y realizar la predicción
for i, edad in enumerate(edades_prediccion):
    # Obtener las características y el valor objetivo para las edades anteriores a la actual
    X = data.loc[data['Age'] < edad, columnas_estadisticas]
    y = data.loc[data['Age'] < edad, columna_valor]
    
    # Entrenar el modelo utilizando los datos anteriores
    modelo.fit(X, y)
    
    # Predecir el valor para las estadísticas actuales y la edad actual
    valor_predicho = modelo.predict([estadisticas_actuales])[0]
    
    # Almacenar el resultado de la predicción en el DataFrame
    resultados_prediccion.loc[i, 'Valor'] = valor_predicho
    
    # Actualizar las estadísticas actuales para la próxima iteración
    if i < len(edades_prediccion) - 1:
        diferencia_edad_actual = estadisticas_promedio.loc[edad] - estadisticas_promedio.loc[edad - 1]
        estadisticas_actuales += diferencia_edad_actual.values.astype(np.float64)

    # Imprimir las características actuales para cada año
    print(f'Características actuales para edad {edad}:')
    print(estadisticas_actuales)
    print()

# Imprimir los resultados de la predicción
print(resultados_prediccion)
import matplotlib.pyplot as plt

# Iterar por cada columna de estadísticas
for columna in columnas_estadisticas:
    # Crear una lista para almacenar los valores de la columna por cada edad
    valores_columna = []
    
    # Obtener los valores de la columna para cada edad
    for edad in edades_prediccion:
        valores_columna.append(estadisticas_actuales[columnas_estadisticas.index(columna)])
        if edad < max(edades_prediccion):
            diferencia_edad_actual = estadisticas_promedio.loc[edad] - estadisticas_promedio.loc[edad - 1]
            estadisticas_actuales += diferencia_edad_actual.values.astype(np.float64)
    
    # Crear una nueva figura para cada columna
    plt.figure()
    
    # Crear el gráfico de línea para la columna
    plt.plot(edades_prediccion, valores_columna)
    
    # Configurar las etiquetas y título del gráfico
    plt.xlabel('Edad')
    plt.ylabel(columna)
    plt.title(f'Cambio de {columna} año tras año')
    
    # Mostrar el gráfico
    plt.show()

```
